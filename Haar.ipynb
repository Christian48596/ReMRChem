```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import cm

def h(x):
    val = 0;
    if x >= 0 and x < 1:
        val = 1
    return val

def w(x):
    return h(2*x)-h(2*x-1)
    
def hnl(x,n,l):
    factor = 2**n
    return np.sqrt(factor) * vh(factor * x - l) 

def wnl(x,n,l):
    factor = 2**n
    return np.sqrt(factor) * vw(factor * x - l) 

def h2(x,y):
    return h(x) * h(y)

def w21(x,y):
    return w(x) * h(y)

def w22(x,y):
    return h(x) * w(y)

def w23(x,y):
    return w(x) * w(y)



vh = np.vectorize(h)
vh2 = np.vectorize(h2)
vw = np.vectorize(w)
vw21 = np.vectorize(w21)
vw22 = np.vectorize(w22)
vw23 = np.vectorize(w23)
```

```python
x = np.arange(-0.2, 1.21, 0.002)
y = wnl(x,1,0)
_ = plt.plot(x,y)
```
